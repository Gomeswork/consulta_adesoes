```python
import requests
import time
import json

import pandas as pd
```

```python
pagina = 1 
uasgs = []
while True:
    data = requests.get(f'https://dadosabertos.compras.gov.br/modulo-uasg/1_consultarUasg?pagina={pagina}&statusUasg=true').json()
    
    uasgs += data['resultado']
    
    if data['paginasRestantes'] == 0:
        break
    time.sleep(1)
    pagina += 1
```

```python
uasgs_df = pd.DataFrame(uasgs)

uasgs_df.drop(columns=[
    'usoSisg', 'adesaoSiasg', 'codigoUnidadePolo', 'nomeUnidadePolo', 'codigoUnidadeEspelho', 'nomeUnidadeEspelho', 'uasgCadastradora', 'cnpjCpfUasg', 'codigoOrgao', 'cnpjCpfOrgao', 'cnpjCpfOrgaoVinculado', 'cnpjCpfOrgaoSuperior', 'codigoSiorg', 'statusUasg', 'dataImplantacaoSidec', 'dataHoraMovimento'
], inplace=True)

uasgs_df
```

```python
with open('uasgs.json', 'w+', encoding='latin-1') as file:
    uasgs_df.to_json(file, indent=4, orient='records', force_ascii=False)
    # json.dump(uasgs_df, file, indent=4, ensure_ascii=False)
```
